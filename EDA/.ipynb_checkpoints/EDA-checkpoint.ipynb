```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import psycopg2
import config
import time

sns.set()
```

### Connect to reviews table for EDA

```python
#connect to AWS RDS
conn = psycopg2.connect(
    host=config.RDS['RDS_HOST'],
    port=config.RDS['RDS_PORT'],
    user=config.RDS['RDS_USER'],
    password=config.RDS['RDS_PASSWORD'],
    dbname=config.RDS['RDS_DBNAME']
)

#establish cursor
cur=conn.cursor()
```

```python
cur.execute('''
    SELECT * FROM ratings
''')

ratings_data = cur.fetchall()
```

```python
ratings_columns = ['rating_id', 'username', 'rating_date', 'rating', 'beer_name',
                  'brewery_name', 'beer_style', 'beer_substyle', 'abv']
```

### Create a pandas DataFrame with the ratings data

```python
#create the DataFrame
ratings = pd.DataFrame(ratings_data, columns=ratings_columns)
#set the index to the rating_id
ratings = ratings.set_index('rating_id')
#convert the rating_date from string to datetime
ratings['rating_date'] = pd.to_datetime(ratings['rating_date'])
```

### Locate and handle missing values

```python
#see if any columns contain null values
ratings.isna().any()

#rating_date has null values
```

```python
ratings.isna().sum()

#ratings_date specifically has null 1,133 null values
```

```python
len(ratings[ratings['rating_date'].isna()])
```

```python
#what percentage of rows will be removed if all null values are removed?
round(len(ratings[ratings['rating_date'].isna()]) / len(ratings), 5)

#0.00026% of the data will be lost, so dropping the null values will not be a problem
```

```python
ratings = ratings.dropna()
ratings.isna().sum()
```
